```python
import csv
from google_play_scraper import Sort, reviews_all

# ID aplikasi yang ingin diambil reviewnya
app_id = "ovo.id"

# Mengambil semua review
reviews = reviews_all(
    app_id,
    sleep_milliseconds=0,  # Tidak ada delay
    lang='id',  # Bahasa Indonesia
    country='id',  # Negara Indonesia
    sort=Sort.NEWEST,  # Urutkan berdasarkan terbaru
    filter_score_with=None  # Ambil semua rating
)

# Menyimpan ke dalam file CSV
csv_filename = "reviews_ovo.csv"

# Menulis ke file CSV
with open(csv_filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Nama", "Rating", "Komentar", "Tanggal"])  # Header CSV

    for review in reviews:
        writer.writerow([review['userName'], review['score'], review['content'], review['at']])

print(f"✅ Data telah disimpan di {csv_filename}")
```
